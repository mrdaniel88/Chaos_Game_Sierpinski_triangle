```python
import random
import matplotlib.pyplot as plt
import numpy as np

# Figure size
plt.rcParams['figure.figsize'] = (12.0, 10.0)

def chaos_game(iterations, random_start):
    #Triangle bounds
    points = ([0,0], [20,40], [40,0])
    #Plot bounds
    plt.scatter([0,20,40], [0,40,0])
    # Define plot limits
    xlim = plt.xlim([-10,50])
    ylim = plt.ylim([-10,50])
    
    # Select randome starting point if there is no starting point as input
    if random_start == None:
        random_x = random.randrange(-10,50,1)
        random_y = random.randrange(-10,50,1)
        random_start = np.array([random_x, random_y])
        print(random_start)
    
    # Plot all points
    for i in range(1,iterations+1):
        plt.scatter(random_start[0],random_start[1], s=1)
        random_point = np.array(random.choice(points))
        random_start = ((random_point-random_start)/2)+random_start
        #print(i)

#chaos_game(iterations = 10000)
        
def iterations_console():
    iterations = int(input("Number of iterations: "))
    random_start = eval(input("Enter the starting point (x,y) or None for a random starting point: "))
    chaos_game(iterations, random_start)
    
iterations_console()
    
    
```
